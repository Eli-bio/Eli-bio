```python
def plot_confusion_matrix(y,y_predict):
  "this function plot_confusion_matrix"
  from sklearn.metrics import confusion_matrix

  cm=confusion_matrix(y,y_predict)
  ax=plt.subplot()
  sns.heatmap(cm,annot=True,ax=ax)
  ax.set_xlabel('Predicted labels')
  ax.set_ylabel('True labels')
  ax.set_title('Confusion Matrix')
  ax.xaxis.set_ticklabels(['did not land','land'])
  ax.yaxis.set_ticklabels(['did not land','land'])
  plt.show()
```

```python
plot_confusion_matrix(Y_test,yhat)
```

```python
#DecisionTree
```

```python
xt_train,xt_test,yt_train,yt_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
tree=DecisionTreeClassifier()
```

```python
patameters={'criterion':['gini','entropy'],
            'splitter':['best','random'],
            'max_depth':[2*n for n in range(1,10)],
            'max_features':['auto','sqrt'],
            'min_samples_leaf':[1,2,4],
            'min_samples_split':[2,5,10]}
```

```python
tree_cv=GridSearchCV(tree,patameters,cv=10)
```

```python
tree_cv.fit(xt_train,yt_train)
```

```python
print("tuned hpyerparameters :(best parameters) ",tree_cv.best_params_)
print("accuracy :",tree_cv.best_score_)
```
